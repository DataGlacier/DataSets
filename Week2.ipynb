### My approach will be first, split the Dataframe, one for Pink Cab and other for Yellow Cab

```python
df_pinkcab = df_cab[df_cab['Company'] == 'Pink Cab']
df_pinkcab
```

```python
df_yellowcab = df_cab[df_cab['Company'] == 'Yellow Cab']
df_yellowcab
```

### Now I'll merge and join the other data sources to each Cab Dataframe before start exploratory analysis

```python
df_tr
```

```python
# Same check for nulls or NaN
df_tr.isnull().sum()
```

```python
df_tr.isna().sum()
```

```python
df_pinkcab = pd.merge(df_pinkcab, df_tr, how='left', on='Transaction ID',validate='one_to_one')
df_pinkcab
```

```python
df_yellowcab = pd.merge(df_yellowcab, df_tr, how='left', on='Transaction ID', validate='one_to_one')
df_yellowcab
```

## In my opinion it can be worth it to join customer information in each of the company dataframes, so I'll do it now

```python
df_cu
```

## All the Data sources are complete, no nulls or NaN observed

```python
# merge customer info to each Cab company dataframe
df_pinkcab = pd.merge(df_pinkcab, df_cu, how='inner', on='Customer ID',validate='m:1')
df_pinkcab
```

```python
df_yellowcab = pd.merge(df_yellowcab, df_cu, how='inner', on='Customer ID', validate='m:1')
df_yellowcab
```
